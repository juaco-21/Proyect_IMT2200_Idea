### Limitaciones y sesgos de los datos y/o analisis.

1. Nos limitamos a extraer informacion de una sola una base de datos, cosa que puede provocar datos falsos poco respaldados o una generalizacion (Sesgo de muestreo).

2. La misma base de datos carece de informacion geografica especifica, cosa que no nos permite analizar mas en profundidad cada localidad de Santiago, las cuales tienden a variar un poco en terminos de temperatura, humedad, etc.

3. Se pueden encontrar ciertas incosistencias en los datos, ya que en algunos años hay ciertos dias que se les registro la temperatura, provocando que en comparacion a los otros años, ciertos analisis estadisticos puede ser menos preciso a causa de la falta de datos.
